```python
import cv2
import numpy as np
#import dlib
import matplotlib.pyplot as plt

from scipy.spatial import distance as dist

from io import BytesIO
from IPython.display import clear_output, Image, display
from PIL import Image as Img

#imagem = cv2.imread("imagens/poligonos.jpg")
imagem = cv2.imread("imagens/mosaico.png")

#imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2HSV)

#Vermelho
limite_inferior = np.array([30, 0, 0])
limite_superior = np.array([255, 0, 0])

#Amarelo
#limite_inferior = np.array([20, 100, 100])
#limite_superior = np.array([30, 255, 255])

#Verde
#limite_inferior = np.array([40, 100, 100])
#limite_superior = np.array([180, 255, 255])

mask = cv2.inRange(imagem, limite_inferior, limite_superior)

result = cv2.bitwise_and(imagem, imagem, mask=mask)

plt.figure(figsize=(20,10))
plt.imshow(result)
```
